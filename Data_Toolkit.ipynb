```python
## QUES 3)  Explain the following terms
# i) The difference in np.array, np.asarray and npYasanyarray.
# ii) The difference between Deep copy and shallow copy
```

i)
1. np.array: This one creates a new array from a given object, like lists or tuples. It has a lot of options for specifying the type and shape of the array. If you give it something that’s already an array, it’ll still create a new one.

2. np.asarray: This guy is similar to np.array but it tries to avoid making a copy if the input is already an array. So, if you pass it an existing array, it’ll just return that array without creating a new one. It’s more efficient when you don’t need a new copy.

3. np.asanyarray: This one is like np.asarray, but it allows for subclasses of arrays too. So, if you have a subclass of an ndarray, it’ll return that subclass instead of converting it to a regular ndarray.

So, in short, if you want a new array, go with np.array. If you want to avoid copies, use np.asarray.

ii)
Here’s the difference between deep copy and shallow copy in a simpler way:

1. Shallow Copy: When you make a shallow copy of an object, you get a new object, but it only copies the references to the original object's elements. So, if the original object has other objects inside it (like lists or dictionaries), both the original and the shallow copy will point to the same inner objects. This means that if you change one of those inner objects in the original, it will also change in the shallow copy.

   You can create a shallow copy in Python using copy() from the copy module or by using the copy() method of a list.

2. Deep Copy: A deep copy, on the other hand, creates a completely new object and also copies all the objects inside it. This means that if the original object has nested objects, the deep copy will have its own copies of those nested objects. So, changes made to the nested objects in the original will not affect the deep copy at all.

   To create a deep copy in Python, you use deepcopy() from the copy module.

In summary:
- Shallow copy shares references, so changes in one affect the other.
- Deep copy creates independent copies, so changes in one do not affect the other.

```python
## QUES 14) Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
# Perform the following tasks using Matplotlib:
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
# right y-axis for 'Humidity').

# b) Label the x-axis as 'Date'.

# c) Set the title of the plot as 'Temperature and Humidity Over Time'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2024-01-01', periods=100, freq='D')

# Create random temperature and humidity data
temperature = np.random.uniform(low=15, high=35, size=100)
humidity = np.random.uniform(low=30, high=90, size=100)

# Create a DataFrame
data = {
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
}
df = pd.DataFrame(data)

# Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the legend
fig.legend(loc='upper right', bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

# Show the plot
plt.show()
```

```python
## QUES 15) Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
# tasks using Matplotlib:
#a) Plot a histogram of the data with 30 bins.

#b) Overlay a line plot representing the normal distribution's probability density function (PDF).

#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

#d) Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Overlay a line plot representing the normal distribution's probability density function (PDF)
mu, std = norm.fit(data)
pdf = norm.pdf(bins, mu, std)
plt.plot(bins, pdf, 'k-', linewidth=2)

# Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show the plot
plt.show()
```

```python
## QUES 17) Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrants
quadrants = []
for i in range(len(x)):
    if x[i] > 0 and y[i] > 0:
        quadrants.append('Quadrant I')
    elif x[i] < 0 and y[i] > 0:
        quadrants.append('Quadrant II')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Quadrant III')
    else:
        quadrants.append('Quadrant IV')

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1')

# Add legend
plt.legend(title='Quadrant')

# Label axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set title
plt.title('Quadrant-wise Scatter Plot')

# Show plot
plt.show()
```

```python
## QUES 19) 8 Using Bo+eh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
# Bar Chart'.
import random
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256

# Enable output in a Jupyter notebook
output_notebook()

# Create random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = [random.randint(1, 100) for _ in categories]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create the figure
# Changed plot_height to height and plot_width to width
p = figure(x_range=categories, height=400, width=600, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Create a color map based on values
color_map = factor_cmap('categories', palette=Viridis256, factors=categories)

# Add bars to the chart
p.vbar(x='categories', top='values', width=0.9, source=source, line_color='white', fill_color=color_map)

# Add hover tool to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show the plot
show(p)
```
