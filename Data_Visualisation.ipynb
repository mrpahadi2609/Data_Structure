```python
## QUES 4) Using the given x and y data, generate a 3D surface plot to visualize the function
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

import plotly.graph_objects as go

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Add title and labels
fig.update_layout(title='3D Surface Plot', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90),
                  scene=dict(
                      xaxis_title='X Axis',
                      yaxis_title='Y Axis',
                      zaxis_title='Z Axis'))

# Show plot
fig.show()
```

```python
## QUES 1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import numpy as np
from bokeh.plotting import figure, show, output_file

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate y values as the sine of x
y = np.sin(x)

# Configure the output to display in a separate HTML file
output_file("sine_wave.html")

# Create a new figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line plot for the sine wave
p.line(x, y, legend_label="y=sin(x)", line_width=2)

# Show the results
show(p)
```

```python
## QUES 4) Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook

# Configure the output to display in the notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad glyph to create the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the results
show(p)
```

```python
## QUES 5) Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.palettes import Viridis256

# Configure the output to display in the notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           toolbar_location=None, tools="")

# Create a color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Add the heatmap to the plot
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the results
show(p)
```
